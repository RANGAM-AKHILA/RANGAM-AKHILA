```python
#copy vs view
import numpy as np
np1=np.array([0,1,2,3,4,5])
#create a view
np2=np1.view()
print(f'original NP1{np1}')
print(f'original NP2{np2}')
np1[0]=41
print(f'changed NP1{np1}')
print(f'original NP2{np2}')
```
